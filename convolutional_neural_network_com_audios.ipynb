<h3>Redes Neurais Convolucionais na Classificação de ÁUDIOS/SONS</h3>

<h4>Instanciação e Treinamento</h4>

```python
from Neuraline.ArtificialIntelligence.DeepLearning.convolutional_neural_network import ConvolutionalNeuralNetwork
from Neuraline.Utilities.chart import Chart
convolutional_neural_network, chart = ConvolutionalNeuralNetwork(), Chart()

convolutional_neural_network.addConvolutionalLayer(feature_maps=[500, 500], max_pooling=False, number_of_cores=5, step=2, activation_function='nonlinear')
convolutional_neural_network.addConvolutionalLayer(feature_maps=[700, 700], max_pooling=True, number_of_cores=10, step=1, activation_function='nonlinear')
convolutional_neural_network.addConvolutionalLayer(feature_maps=[500, 500], max_pooling=False, number_of_cores=5, step=2, activation_function='nonlinear')

result1 = convolutional_neural_network.addFit(url_path='./audios/training/homem.wav', class_name='homem')
result2 = convolutional_neural_network.addFit(url_path='./audios/training/mulher.wav', class_name='mulher')
if result1 and result2: print('treinamento concluído com sucesso')
else: print('erro ao concluir o treinamento')
```

<h4>Teste com ÁUDIO/VOZ Masculino(a)</h4>

```python
url_inputs = './audios/test/homem.wav'
chart.plotAUDIO(url_inputs)
classification = convolutional_neural_network.predict(url_inputs)
print(f'resultado classificativo para voz de {classification}')
```

<h4>Teste com ÁUDIO/VOZ Feminino(a)</h4>

```python
url_inputs = './audios/test/mulher.wav'
chart.plotAUDIO(url_inputs)
classification = convolutional_neural_network.predict(url_inputs)
print(f'resultado classificativo para voz de {classification}')
```
